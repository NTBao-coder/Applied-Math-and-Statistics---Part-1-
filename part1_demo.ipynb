# Part 1 Demo
Notebook này demo 1 ma trận bất kỳ và các hàm xử lý trong `part1.py`.

```python
from part1 import (
    gaussian_elimination, backward_elimination, solver, determinant, inverse, rank_and_basis, verify_solution
)
# Ma trận demo
A = [
    [2.0, -1.0, 1.0],
    [3.0, 3.0, 9.0],
    [3.0, 3.0, 5.0],
]
b = [2.0, -1.0, 4.0]
```

Ma trận A và vế phải b:

```python
print('A =')
for row in A:
    print(row)
print('b =', b)
```

1. Khử Gauss

```python
M_ref, s, pivot_cols = gaussian_elimination([row[:] for row in A], b[:])
print('REF (A|b):')
for row in M_ref:
    print([round(x, 6) for x in row])
print('Số lần đổi dòng: ', s)
print('Danh sách các cột có chứa pivot: ', pivot_cols)
```

2. Hàm khử ngược

```python
M_rref = backward_elimination(M_ref, pivot_cols)
print('RREF (A|b):')
for row in M_rref:
    print([round(x, 4) for x in row])
```

3. Giải hệ Ax = b

```python
status, M, res, swap_count = solver([row[:] for row in A], b[:])
print('Trạng thái:', status)
print('Số lần đổi dòng:', swap_count)
print('Nghiệm của hệ:', res)
```

4. Định thức và nghịch đảo của ma trận

```python
detA = determinant([row[:] for row in A])
invA = inverse([row[:] for row in A])
print('Giá trị định thức của ma trận: ', detA)
print('Ma trận nghịch đảo: ')
if isinstance(invA, list):
    for row in invA:
        print([round(x, 6) for x in row])
else:
    print(invA)
```

5. Hạng và cơ sở (cơ sở dòng, cột, cơ sở nghiệm)

```python
rb = rank_and_basis([row[:] for row in A])
print('Hạng của ma trận: ', rb['rank'])
print('Cơ sở dòng: ', rb['row_basis'])
print('Cơ sở cột: ', rb['col_basis'])
print('Cơ sở không gian nghiệm: ', rb['null_basis'])
```

6. So sánh nghiệm với Numpy

```python
if isinstance(res, list):
    print('So sánh: ', verify_solution(A, res, b))
elif isinstance(res, dict):
    print('So sánh ', verify_solution(A, res['particular'], b))
else:
    print('Phương trình vô nghiệm nên không thể so sánh với Numpy')
```
